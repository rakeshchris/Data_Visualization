*Removing the NULL Values*

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

*Checking for Duplicates*

```python
print(df.duplicated())
```

```python
df.info()
```

*Converting the 'DATE_DIED' column to 'DIED' and Changing values to 0 and 1 where 0=died and 1=Alive*

```python
df['DATE_DIED']=np.where(df['DATE_DIED']=='9999-99-99',1,0)
```

```python
df.rename(columns = {'DATE_DIED':'DIED'}, inplace = True)
```

```python
df
```

*Changing the data type of column 'DIED' from object to int*

```python
df['DIED']=pd.to_numeric(df['DIED'])
```

```python
df.dtypes
```

# Data Visualization

*By using the Pandas cleaning the data ,removing null values,duplicates or any unwanted data has been completed*
*Now the data is visualized in from of graphs or plots using matplot or seaborn*

```python
df['AGE'].plot(kind='hist',color='blue')                   #Checking different age persons present in the data 
```

```python
sns.countplot(data=df,x='DIABETES',color='red')
```

```python
sns.countplot(data=df,x='DIED',COLOR='PURPLE')
```

```python
 sns.displot(data=df,x='AGE',hue='SEX',col='DIED',multiple='stack', palette = 'bright') 
                                                
```

    *We can see the persons died due to covid with respective to their sex*

```python
sns.boxplot(data=df,y='AGE',x='TOBACCO',hue='SEX').set_xlim(0, 4)    
```

     *Boxplot b/w age and tobacco with respective to sex*

```python
sns.displot(data=df,x='AGE',hue='SEX',col='PATIENT_TYPE',multiple='stack',palette='dark')
```

```python
corr=df.corr()
sns.heatmap(corr,linewidths=0.5,linecolor='black')
```

       *Correlation:Relationship between each column in the data set is stored in corr*
          *Heat map is presented using correaltion*
